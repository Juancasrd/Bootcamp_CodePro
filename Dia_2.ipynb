```python
edad_festival = int(input('Que edad tiene?'))

if edad_festival < 18:
    print('Acceso denegado - Solo Zona Kids.')

elif edad_festival < 21:
    print('Acceso permitido - No puede comprar bebidas alcoholicas.')

else:
    print('Acceso total - Disfrute el festival.')
```

```python
temperatura = int(input('Cuantos grados hace?'))

if temperatura >= -10 and temperatura <= 13:
    print("Ro'y koko")

elif temperatura >= 14 and temperatura <= 23:
    print("Refresca'i")

elif temperatura >= 24 and temperatura <= 28:
    print("Oikote")

elif temperatura >= 29 and temperatura <= 45:
    print("Hakurooooo")

else:
    print("Ñamano mba'e!")
```
