```python
cantidad_de_caramelos = int(input('Cuantos caramelos hay?'))
cantidad_de_ninhos = int(input('Cuantos ninhos hay?'))

cantidad_por_ninho = cantidad_de_caramelos / cantidad_de_ninhos
cantidad_que_sobro = cantidad_de_caramelos & cantidad_de_ninhos

print(f'cada ninho recibio {cantidad_por_ninho}.')
print(f'sobraron {cantidad_que_sobro} caramelos.')
```
